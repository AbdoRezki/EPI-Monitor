```python
import os
import shutil

# Define the path to your dataset
dataset_path = '/kaggle/input/helmet-data/data'
image_data_path = os.path.join(dataset_path, 'images')
labels_data_path = os.path.join(dataset_path, 'labels')
# Create a directory for the images and labels in the working directory
output_path = '/kaggle/working/data'
image_path = os.path.join(output_path, 'images')
os.makedirs(image_path, exist_ok=True)
label_path = os.path.join(output_path, 'labels')
os.makedirs(label_path, exist_ok=True)
# If the file is an image, copy it to the images directory
for filename in os.listdir(image_data_path):
    src_path = os.path.join(image_data_path, filename)
    dst_path = os.path.join(image_path, filename)
    shutil.copy(src_path, dst_path)
# If the file is a label, copy it to the labels directory
for filename in os.listdir(labels_data_path):
    src_path = os.path.join(labels_data_path, filename)
    dst_path = os.path.join(label_path, filename)
    shutil.copy(src_path, dst_path)
```
